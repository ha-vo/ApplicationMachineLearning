````python
from sklearn.model_selection import KFold
kf = KFold(n_splits=50, shuffle = True)
````

````python
X = data.iloc[:,:11]
y = data.iloc[:,11]
````

````python
for train_index, test_index in kf.split(X):
  print("Train:", train_index, "Test:", test_index)
  X_train, X_test = X.iloc[train_index], X.iloc[test_index]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]
  print("X test", X_test)
  print("==============")
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
````

f. Tính độ chính xác tổng thể cho mỗi lần lặp và độ chính xác tổng thể của trung bình
50 lần lặp

````python
sum = 0
for i in range(10, 61):
  kf = KFold(n_splits=i, shuffle = True)
  X1 = data.iloc[:,:11]
  y1 = data.iloc[:,11] 
  for train_index, test_index in kf.split(X1):
    X_train1, X_test1 = X1.iloc[train_index], X1.iloc[test_index]
    y_train1, y_test1 = y1.iloc[train_index], y1.iloc[test_index]
  clf = DecisionTreeClassifier(criterion="gini", random_state=100, max_depth = 5,min_samples_leaf=5)
  clf.fit(X_train1, y_train1)
  y_pred1 = clf.predict(X_test1)
  sum += accuracy_score(y_test1, y_pred1)*100
sum / 50
````

Trung bình độ chính xác trong 50 lần lặp là: 59.1
